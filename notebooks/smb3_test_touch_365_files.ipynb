## Touch 365 files on SMB3 mounted share via Wifi

```python
import pandas as pd
from pathlib import Path

# Get os separator
sep = os.sep

# Set the source directory for the test (SMB3 mounted share)
base_dir = "/Volumes/home/test_src"

# Build a pandas datetime index of 365 days
dates = pd.date_range(start="2016-01-01", end="2017-01-01")

# COnvert timestamps to strings 
dates_str = dates.strftime("%Y-%m-%d")

# Touch file with name "YYYY-MM-DD_test-file.txt"
for ds in dates_str:
    Path(f"{base_dir}{sep}{ds}_test-file.txt").touch()
```
